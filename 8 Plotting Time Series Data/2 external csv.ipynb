```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('ggplot')

data = pd.read_csv('data.csv')

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

# make date visiable properly by slanting them by modifying current formatter
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```

```python
plt.style.use('ggplot')

data = pd.read_csv('data.csv')

# the dates from data.csv are in string so we have to convert it to date type for accurate plotting like if the old date is at last then it will not be displayed at front if its data type is string. Use sort for correct plotting
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

# make date visiable properly by slanting them by modifying current formatter
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```
