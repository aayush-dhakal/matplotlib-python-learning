```python
import pandas as pd
from datetime import datetime, timedelta # python native module
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.style.use('ggplot')

plt.plot_date(dates, y)
```

```python
plt.plot_date(dates, y, linestyle='solid')
```

```python
plt.plot_date(dates, y, linestyle='solid')

# make date visiable properly by slanting them by modifying current formatter
plt.gcf().autofmt_xdate()
```

```python
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

# format how the dates are displayed
date_format = mpl_dates.DateFormatter('%b, %d %Y')

# format the current axis
plt.gca().xaxis.set_major_formatter(date_format)
```
